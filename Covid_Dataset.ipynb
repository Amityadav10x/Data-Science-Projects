**Correlation Analysis**

```python

# Selecting relevant columns for correlation analysis

correlation_data = df[['COVID-19 deaths per 100,000 population in 2020',
                       'Gini coefficient of income',
                       'All-cause death count, 2017', 'All-cause death count, 2018',
                       'All-cause death count, 2019', 'All-cause death count, 2020',
                       'Excess deaths in 2020 per 100,000 population, relative to 2017-2019 average']]

# Calculating Pearson correlation coefficients
pearson_corr = correlation_data.corr()

# Calculating Spearman rank correlation coefficients
spearman_corr = correlation_data.corr(method='spearman')

# Displaying the correlation matrices
print("Pearson Correlation:")
print(pearson_corr)

print("\nSpearman Rank Correlation:")
print(spearman_corr)
```

**Death Rate per 1000 People (2019)**

```python
selected_countries = ["Morocco", "United Kingdom", "Spain", "Poland","Colombia"]
filtered_df = df[df['Country name'].isin(selected_countries)].copy()

filtered_df.loc[:, 'All-cause death count, 2019'] = filtered_df['All-cause death count, 2019'].fillna(0)

mlist = filtered_df['All-cause death count, 2019']
population = filtered_df['Population 2019']
labels = filtered_df['Country name']

death_rate_per_capita = (mlist / population) * 1000


plt.figure(figsize=(10, 6))
plt.bar(labels, death_rate_per_capita)
plt.xlabel('Country',color='gray')
plt.ylabel('Death Rate per 1000 People')
plt.title('Death Rate per 1000 People (2019)')
plt.xticks(rotation=45)
plt.show()
```

**Death Rate per 1000 People (2020)**

```python
selected_countries = ["India", "Germany", "Japan", "France", "Colombia","Hungary"]
filtered_df = df[df['Country name'].isin(selected_countries)].copy()

filtered_df.loc[:, 'All-cause death count, 2017'] = filtered_df['All-cause death count, 2017'].fillna(0)

mlist = filtered_df['All-cause death count, 2017']
population = filtered_df['Population 2020']
labels = filtered_df['Country name']

death_rate_per_capita = (mlist / population) * 1000


plt.figure(figsize=(10, 6))
plt.bar(labels, death_rate_per_capita)
plt.xlabel('Country')
plt.ylabel('Death Rate per 1000 People')
plt.title('Death Rate per 1000 People (2020)')
plt.xticks(rotation=45)
plt.show()
```
