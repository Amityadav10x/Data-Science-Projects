# **Customer Behavior**

**Distribution of Booking Pattern**

```python
import matplotlib.pyplot as plt


# Assuming your dataset is stored in a pandas DataFrame named 'df'

# Create a new column for total stays
df['total_stays'] = df['stays_in_weekend_nights'] + df['stays_in_week_nights']

# Group by total stays and count the occurrences
booking_patterns = df['total_stays'].value_counts().reset_index()
booking_patterns.columns = ['total_stays', 'count']

# Sort the DataFrame by total stays
booking_patterns = booking_patterns.sort_values(by='total_stays')

# Set a color gradient
colors = sns.color_palette("viridis", len(booking_patterns))

# Plotting the most common booking patterns

sns.set(style="whitegrid")
plt.figure(figsize=(14, 8))
sns.barplot(x='total_stays', y='count', data=booking_patterns, palette=colors)
plt.title('Distribution of Booking Patterns', fontsize=16)
plt.xlabel('Total Stays (Weekend Nights + Week Nights)', fontsize=14)
plt.ylabel('Number of Bookings', fontsize=14)

# Adding labels to each bar
for index, value in enumerate(booking_patterns['count']):
    plt.text(index, value + 1, str(value), ha='center', va='bottom', fontsize=10, color='black')

plt.show()
```

**Checking for coustomer behaviour over the years**

```python
# Group by arrival_date_year and calculate average lead time and cancellation rate

yearly_stats = df.groupby('arrival_date_year').agg({'lead_time': 'mean', 'is_canceled': 'mean'}).reset_index()

# Plotting the trends
sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))

# Plot average lead time
plt.subplot(2, 1, 1)
sns.lineplot(x='arrival_date_year', y='lead_time', data=yearly_stats, marker='o', color='skyblue')
plt.title('Average Lead Time Over the Years')
plt.xlabel('Year')
plt.ylabel('Average Lead Time (days)')

# Plot cancellation rate
plt.subplot(2, 1, 2)
sns.lineplot(x='arrival_date_year', y='is_canceled', data=yearly_stats, marker='o', color='coral')
plt.title('Cancellation Rate Over the Years')
plt.xlabel('Year')
plt.ylabel('Cancellation Rate')

plt.tight_layout()
plt.show()
```

# **Cancellation Analysis**

**Overall canselation rate**

```python
overall_cancellation_rate = df['is_canceled'].mean()

# Plotting the cancellation rate
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))

# Plot a bar for overall cancellation rate
sns.barplot(x=['Overall Cancellation Rate'], y=[overall_cancellation_rate], color='coral')
plt.title('Overall Cancellation Rate')
plt.ylabel('Cancellation Rate')
plt.ylim(0, 1)  # Set y-axis limit to represent percentage

# Adding the cancellation rate value on top of the bar
plt.text(0, overall_cancellation_rate + 0.02, f'{overall_cancellation_rate:.2%}', ha='center', va='bottom', fontsize=12)

plt.show()
```

 **Identify relevant metrics for comparison (e.g., lead_time, booking_changes, total_of_special_requests)**

```python

# Spliting the dataset into repeated guests and new guests
repeated_guests = df[df['is_repeated_guest'] == 1]
new_guests = df[df['is_repeated_guest'] == 0]

# Calculating average values for each group
avg_lead_time_repeated = repeated_guests['lead_time'].mean()
avg_lead_time_new = new_guests['lead_time'].mean()


avg_changes_repeated = repeated_guests['booking_changes'].mean()
avg_changes_new = new_guests['booking_changes'].mean()


avg_special_requests_repeated = repeated_guests['total_of_special_requests'].mean()
avg_special_requests_new = new_guests['total_of_special_requests'].mean()


# Printing the results

print(f"Average Lead Time for Repeated Guests: {avg_lead_time_repeated}")
print(f"Average Lead Time for New Guests: {avg_lead_time_new}")

print(f"Average Booking Changes for Repeated Guests: {avg_changes_repeated}")
print(f"Average Booking Changes for New Guests: {avg_changes_new}")

print(f"Average Special Requests for Repeated Guests: {avg_special_requests_repeated}")
print(f"Average Special Requests for New Guests: {avg_special_requests_new}")
```

# **Previous Booking History**

```python
# Creating a new column to categorize guests based on previous cancellations

df['previous_cancellation_status'] = df['previous_cancellations'].apply(lambda x: 'No Cancellation' if x == 0 else 'Cancellations')

# Calculating the counts for each category
category_counts = df.groupby(['previous_cancellation_status', 'is_canceled']).size().unstack(fill_value=0)

# Ploting the stacked bar chart
category_counts.plot(kind='bar', stacked=True, color=['green', 'red'], figsize=(10, 6))

plt.title('Impact of Previous Cancellations on Current Bookings')
plt.xlabel('Previous Cancellation Status')
plt.ylabel('Count')
plt.legend(title='Current Booking Status', labels=['Not Canceled', 'Canceled'])
plt.show()

```
