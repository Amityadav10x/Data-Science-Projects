```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

**Spilliting the dataset into features and target variable**

```python
x = data[['mass','width','height','color_score']]
y = data['fruit_label']
```

```python
x.head(5)
```

```python
y.head(5)
```

```python
x_train, x_test, y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=1)

scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
x_train.head(5)
```

```python
x_test.head(5)
```

```python
k = 5
knn_classifier = KNeighborsClassifier(n_neighbors=k)
```

```python


knn_classifier.fit(x_train_scaled,y_train)
y_pred = knn_classifier.predict(x_test_scaled)

```

```python
accuracy  = accuracy_score(y_test,y_pred)
```

```python
accuracy
```

```python
pca = PCA(n_components=2)
x_test_pca = pca.fit_transform(x_test_scaled)
```

```python
plt.figure(figsize=(10,6))

plt.scatter(x_test_pca[:,0],x_test_pca[:,1],c=y_test,cmap='viridis',marker='o',label='Actual')
plt.scatter(x_test_pca[:,0],x_test_pca[:,1],c=y_pred,cmap='jet', marker='x',label='Predicted', alpha=0.5)

plt.title('Actual vs predicted labels (k -nn)')
plt.xlabel('principle component 1')
plt.ylabel('pricniple component 2')
plt.legend()
plt.colorbar(label='Fruit Label')
plt.grid(True)
plt.show()
```
