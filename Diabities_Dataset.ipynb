#                                          **Emergency and Outpatient Visits**

```python
patients_with_visits = df[(df['number_emergency'] > 0) | (df['number_outpatient'] > 0)]

# Counting the number of unique patients visits

num_patients_with_visits = patients_with_visits['patient_nbr'].nunique()

print(f"The number of patients with emergency or outpatient visits: {num_patients_with_visits}")
```

```python
# Selecting relevant columns for analysis
columns_of_interest = ['number_emergency', 'number_outpatient', 'time_in_hospital', 'num_medications', 'num_procedures', 'num_lab_procedures']

# Creating a correlation matrix
correlation_matrix = df[columns_of_interest].corr()

# Creating a heatmap to visualize correlations
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=1)
plt.title('Correlation Matrix: Emergency/Outpatient Visits and Other Factors')
plt.show()


# Note-

# Values close to 1 indicate a strong positive correlation.
# Values close to -1 indicate a strong negative correlation.
# Values close to 0 indicate a weak or no correlation.
```

# **Analyzing Changes in Medication**







```python
# Counting the occurrences of 'Ch' in the 'change' column

change_count = df['change'].value_counts().get('Ch', 0)

# Calculating the percentage of patients with a change in medication

total_patients = len(df)
percentage_change = (change_count / total_patients) * 100

print(f"The percentage of patients with a change in medication: {percentage_change:.2f}%")
```

**Checking correlation between changes in medication and readmission rates**

```python
# Creating a DataFrame with relevant columns

columns_of_interest = ['change', 'diabatic', 'time_in_hospital', 'num_lab_procedures', 'num_procedures', 'num_medications']
data_for_correlation = df[columns_of_interest]

# Mapping categorical values to numericaldata for correlation analysis

data_for_correlation['change'] = data_for_correlation['change'].map({'No': 0, 'Ch': 1})
data_for_correlation['diabatic'] = data_for_correlation['diabatic'].map({'No': 0, 'Yes': 1})

# Creating a correlation matrix

correlation_matrix = data_for_correlation.corr()

# Creating a heatmap to visualize correlations

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=1)
plt.title('Correlation Matrix: Changes in Medication and Readmission Rates')
plt.show()


# note:-

# Values range from -1 to 1:

# -1 indicates a perfect negative correlation.
# 1 indicates a perfect positive correlation.
# 0 indicates no correlation.
```
