```python
data = {
    'fruit_label': [1, 1, 1, 2, 2, 2, 2, 2, 1, 1],
    'fruit_name': ['apple', 'apple', 'apple', 'mandarin', 'mandarin', 'mandarin', 'mandarin', 'mandarin', 'apple', 'apple'],
    'fruit_subtype': ['granny_smith', 'granny_smith', 'granny_smith', 'mandarin', 'mandarin', 'mandarin', 'mandarin', 'mandarin', 'braeburn', 'braeburn'],
    'mass': [192, 180, 176, 86, 84, 80, 80, 76, 178, 172],
    'width': [8.4, 8.0, 7.4, 6.2, 6.0, 5.8, 5.9, 5.8, 7.1, 7.4],
    'height': [7.3, 6.8, 7.2, 4.7, 4.6, 4.3, 4.3, 4.0, 7.8, 7.0],
    'color_score': [0.55, 0.59, 0.60, 0.80, 0.79, 0.77, 0.81, 0.81, 0.92, 0.89]
}

df = pd.DataFrame(data)
```

 **Visualizaing previous data**


```python

plt.hist(df['mass'], bins=20, color='blue', edgecolor='black')

sns.set(style="darkgrid")

plt.xlabel('Mass')
plt.ylabel('Frequency')
plt.title('Distribution of Mass in the Dataset')

plt.show()
```

# **Plotly questions**

**Interactive scatter plot**

```python
import plotly.express as px

fig = px.scatter(df,x='width', y='mass', color='fruit_name', size='mass',
                 hover_data=['fruit_subtype'], title='Scatter Plot: Diameter vs Mass',
                 labels={'width': 'Diameter', 'mass': 'Mass'})

fig.show()
```

# **Pie chart**

**Checking Distribution of all fruits**

```python
fruit_counts = df['fruit_name'].value_counts().reset_index()

fruit_counts.columns = ['fruit_name', 'count']

fig = px.pie(fruit_counts, names='fruit_name',title='Fruit Distribution in the Dataset')


fig.show()
```

# **3D scatter plot**

**Ploted a 3d scatter plot with mass height and diameter**

```python
fig = px.scatter_3d(df, x='width', y='mass', z='height', color='fruit_name',
                    size='mass', hover_data=['fruit_subtype'],
                    title='3D Scatter Plot: Mass, Diameter, and Height',
                    labels={'width': 'Diameter', 'mass': 'Mass', 'height': 'Height'})
fig.show()
```

# **Box Plot**

**Distribution of mass for each fruit**

```python
fig = px.box(df, x='fruit_name', y='mass', color='fruit_name',

             title='Box Plot: Distribution of Mass for Each Fruit',
             labels={'mass': 'Mass', 'fruit_name': 'Fruit Name'})

fig.show()
```

# **Matplotlib**

```python
# Grouping the data by 'fruit_name' and calculating the average mass
average_mass_by_fruit = df.groupby('fruit_name')['mass'].mean().reset_index()

# Plotting the bar chart

plt.figure(figsize=(15, 5))
plt.bar(average_mass_by_fruit['fruit_name'], average_mass_by_fruit['mass'], color='hotpink')
plt.xlabel('Fruit Name')
plt.ylabel('Average Mass')
plt.title('Average Mass for Each Fruit')
plt.show()
```

**Plotting the trends of diameter using line plot**

```python
# Plotting the trends of diameter

plt.figure(figsize=(15, 5))
plt.plot(df['width'], marker='o', linestyle='-', color='blue')
plt.xlabel('Index')
plt.ylabel('Width (Diameter)')
plt.title('Trends of Width (Diameter) over the Dataset')
plt.grid(True)
plt.show()
```

**Creating subplots to display side by side box plot for mass diameter and height**

```python
# Creating a subplot with 1 row and 3 columns

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Box plot for 'mass'
axes[0].boxplot(df['mass'])
axes[0].set_title('Box Plot: Mass')

# Box plot for 'width' (diameter)

axes[1].boxplot(df['width'])
axes[1].set_title('Box Plot: Diameter')

# Box plot for 'height'

axes[2].boxplot(df['height'])
axes[2].set_title('Box Plot: Height')

# Adding labels to the y-axes
for ax in axes:
    ax.set_ylabel('Values')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
